```python
## Daniel Kim
## CS 325
## Winter 2020
## This class is so harddddd D:
## HW8

# Bin Packing Approximations
# 
# 1. , first_fit, first_fit_decreasing, best_fit
# Used to pack bins
# 2.createRandom
# Creates random bin data file called "random_bin.txt"
# 3.Helper Functions:  menu, drive_main, loadFile
# Provides menu, drives main file, and loads data files respectively
###

from collections import defaultdict 
import queue
import sys

import random

def createRandom():
    with open("random_bin.txt", "w") as f:
        case_count = 300
        f.write("{}\n".format(case_count))
        for i in range(case_count):
            bin_capacity = random.randint(10, 50)
            number_of_weights = random.randint(10, 30)
            item_weights = [str(random.randint(2, bin_capacity))
                       for _ in range(number_of_weights)]
            f.write("{}\n".format(number_of_weights))
            f.write("{}\n".format(bin_capacity))
            f.write("{}\n".format(" ".join(item_weights)))

def loadFile(name):
    test_cases = []
    with open(name, 'r') as file:
        test_count = int(file.readline())
        for i in range(test_count):
            bin_capacity = int(file.readline())
            item_count = file.readline()
            items = [int(x) for x in file.readline().strip().split(' ')]
            test_cases.append({
                "capacity": bin_capacity,
                "items": items,
            })
    return test_cases


def first_fit(itemWeights, capacity):
    bins_list = []
    for weight in itemWeights:
        # bin is a reserved keyword
        found_bin = False
        # print(f"Bins: {bins}")
        for i in range(len(bins_list)):
            # print(f"Bin: {bn}, Weight: {weight}")
            if bins_list[i] >= weight:
                bins_list[i] -= weight
                found_bin = True
                break     
        if not found_bin:
            bins_list.append(capacity - weight)
            
    bin_length = len(bins_list)
    return bin_length


def first_fit_decreasing(itemWeights, capacity):
    itemWeights.sort(reverse=True)
    return first_fit(itemWeights, capacity)


def best_fit(itemWeights, capacity):
    bins_list = []
    for weight in itemWeights:
        remaining_cap = capacity + 1
        index = 0
        for i in range(len(bins_list)):
            if bins_list[i] >= weight and bins_list[i] - weight < remaining_cap:
                index = i
                remaining_cap = bins_list[i] - weight
        if remaining_cap == capacity + 1:
            bins_list.append(capacity - weight)
        else:
            bins_list[index] -= weight
    bin_length = len(bins_list)
    
    return bin_length

def menu():
    status = True
    while status:
        print("----ENTER 1 to Use bin.txt")
        print("----ENTER 2 to Use randomly generated bin data")
        print("Please enter 1 or 2")
        value = input()
        try:
            answer = int(value)
            if answer == 1 or answer == 2:
                break
            else:
                continue
        except ValueError:
             print ("This is not a valid value")
            
    if answer == 1:
        fileName = "bin.txt"
        data1 = loadFile(fileName)
        return data1
    elif answer == 2:
        fileName = "random_bin.txt"
        data2 = loadFile(fileName)
        return data2


def drive_main(binData):
    summary_results = {}
    for caseNum, case in enumerate(binData):
        items = case['items']
        bin_capacity = case['capacity']
        firstFit = first_fit(items, bin_capacity)
        bestFit = best_fit(items, bin_capacity)
        firstDecrease = first_fit_decreasing(items, bin_capacity)
        results = { "first-fit": firstFit, "first-fit-decreasing": firstDecrease,  "best_fit": bestFit,
        }

        optimalResult = min(results, key=results.get)
        
        if firstFit == firstDecrease and firstDecrease == bestFit:
            optimalResult = "All methods had equal performance"

        print("-----------------------------")
        print("[Test Case {}] First-Fit: {}, First-Fit-Decreasing: {}, Best Fit: {} \n  *Optimal Method: {}".format(
            caseNum + 1, firstFit, firstDecrease, bestFit, optimalResult))
        
        if optimalResult in summary_results:
            summary_results[optimalResult] += 1
        else:
            summary_results[optimalResult] = 1

    for result in summary_results:
        percentage = round(summary_results[result] / len(binData), 4) * 100
        print("The packing method [ {} ] resulted in the best outcome [ {}% of iterations ]".format(
            result, percentage))
    print("--------------End of Data------------------")
    
def main():
    #creates random numbers for bin data
    createRandom()
    data = menu()
    #Driver function
    drive_main(data) 

main()
```
