```python
## Daniel Kim
## CS 325
## Winter 2020
## This class is so harddddd
## HW4

# Program for Make Change problem
# Functions used:
# 1. get_denominations 
# returns list of calculated denomations for c and k
# 2. count_frequency
# returns frequency of each denomination used
# 3. find_change
# returns a list of denominations used to make change
# 4. make_change_driver
# reads data.txt and writes change.txt
# Driver for the first 3 functions


#helper function
#calculates the denomination values for c and k
def get_denominations(c, k):
    denomination = []
    # Calculated values of c^0, c^1 ..c^k
    for i in range(k + 1):
        denomination.append(pow(c,i))
    return denomination

#helper function 
#counts frequency of each denomination
def count_frequency(coin_results):
    # Dictionary to count denomination frequency  
    coin_frequency = {}
    for c in coin_results:
        if c in coin_frequency:
            coin_frequency[c] += 1
        else:
            coin_frequency[c] = 1
    return coin_frequency

# Logs finding the change in coins!
def find_change(Total, denominations):
    # logs the denomination used for each subtraction from N
    my_change = []
    i = len(denominations) - 1
    while i >= 0:
        # times a denomination value can divide into N
        # denomination cant be greater than current subtotal
        while Total >= denominations[i]:
            Total -= denominations[i]
#             print("coin value", denominations[i])
#             print("Total Sum Left", Total)
            my_change.append(denominations[i])
        i -= 1
    return my_change

def make_change_driver(fileName, writeFile):
    # reading from the data.txt
    file_contents = open(fileName, 'r')
    
    #write to file
    file_write = open(writeFile, 'w')
    lines = file_contents.readlines()

    # reading data line by line
    for line in lines:
        # splitting the value
        val = line.strip().split(" ")
        print(val)
        # creating the denomination array
        d = get_denominations(int(val[0]), int(val[1]))
        coins_change = count_frequency(find_change(int(val[2]), d))
        for key, val in coins_change.items():
            # saving the key value pair in file
            print("Output:", key, val)
            file_write.write(str("%d %d"%(key, val)) + "\n")
        file_write.write("\n")
    # closing the file after saving the result
    file_write.close()
    print("----- Write to file complete -------")



# Driver program for testing knapsack_BU function 
def main():
    filename = "data.txt"
    writeFile = "change.txt"
    
    print("-------makeChange Algorithm-------------")
    
    print("-Sets of [c =denomination, k=kth power of c, n =sum total] -------------")
    make_change_driver(filename, writeFile)
    
main()



```
