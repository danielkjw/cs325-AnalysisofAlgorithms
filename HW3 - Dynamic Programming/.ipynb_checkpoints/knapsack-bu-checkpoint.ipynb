```python
## Daniel Kim
## CS 325
## Winter 2020
## This class is so harddddd
## HW4 - Knapsack Bottom-Up

# Program for 0-1 Knapsack problem 
# Returns:
# 1. V[n][W] = Maximum value that can be put in a knapsack of capacity W 
# 2. Array V with dimensions n+1 by W+1 
# Parameters:
# W = capacity of knapsack
# n = length of items value array or weight array
# weight = contains item weights
# value = contains item value respective to weight index

def knapsack_BU(W, weight, value, n):
    cols = W + 1
    rows = n + 1
    #V is a 2D array with [values][weight]
    V = [[0 for x in range(cols)] for x in range(rows)] 
    for i in range(rows): 
        for w in range(cols): 
            if i == 0 or w == 0: 
                 V[i][w] = 0
            elif weight[i-1] <= w: 
                V[i][w] = max( V[i-1][w-weight[i-1]] + value[i-1],  V[i-1][w]) 
            else: 
                V[i][w] = V[i-1][w]
    #---end of calculations---                  
    return V[n][W], V

# Driver program for testing knapsack_BU function 
def main():
    filename = "data.txt"
    bweight = []
    cvalue = []
    

    W = 6
    weight = [] 
    value = []

    print("-------knapsack Algorithm-------------")
    
    with open(filename, "r") as file:
        newline = file.readlines()
        print(newline[0])

        c_list = map(int, newline[0].split(','))
        c_list2 = map(int, newline[1].split(','))

        for i in c_list:
            bweight.append(i)
        for i in c_list2:
            cvalue.append(i)

        weight = bweight

        
        value = cvalue
        n = len(value)

        print("-------Weight Array-------------")          
        print(weight)
        len(weight)
        print("-------Value Array-------------")
        print(value)
        
        #edit function parameters here
        print("------------------")
        W = 6
        # function returns V[n][W], Array V respectively
        val, V = knapsack_BU(W, weight, value, n)
        #printing solution
        solution = V[n][W]
        weight_col = W
        item_list = []
        val_list = []
        #countdown from n to 0
        for i in range(n,0,-1):
            if (solution < 0):
                break
            #trace back value starting from top or final value of table    
            if (solution == V[i-1][weight_col]):
                continue
            else:
                item_list.append(weight[i - 1])
                val_list.append(value[i-1])
                solution = solution - value[i - 1]
                weight_col = weight_col - weight[i-1]
    
        print(f"Items are: weights {item_list} with respective values of {val_list}") 
        print(f"Maximal value is: {val}")


main()



```
